```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"Base_de_Transacoes_e_Cupons_Capturados.csv", sep=";")

df.head(3)
```

```python
# Contar quantos cupons cada loja teve
cupons_por_loja = df.groupby("nome_estabelecimento")["id_cupom"].nunique()

# Calcular a média e o desvio padrão
media = cupons_por_loja.mean()
desvio = cupons_por_loja.std(ddof=1)

# Calcular o erro padrão
n = len(cupons_por_loja)
erro_padrao = desvio / np.sqrt(n)

# Calcular intervalo de confiança de 95%
Z = 1.96  # 95%
ic_inferior = media - Z * erro_padrao
ic_superior = media + Z * erro_padrao

print(f"Média de cupons por loja: {media:.2f}")
print(f"IC = [{ic_inferior:.2f}, {ic_superior:.2f}]")

print(cupons_por_loja)
```
